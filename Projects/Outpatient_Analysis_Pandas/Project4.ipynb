```python
import pandas as pd

# Load data
opdf = pd.read_csv('./outpatient_sample.csv')

opdf.head()
```

# Calculate Total Visits

```python
# Calculate the number of unique visits for each patient
total_visits = opdf.groupby('Patient_ID')['Visit_ID'].nunique()

# Convert to dataframe
total_visits_df = total_visits.reset_index()
total_visits_df.columns = ['Patient_ID', 'Total_Visits']

print(total_visits_df)
```

# Total unique physicians seen

```python
# Combine physician columns to calculate total unique physicians per patient
total_physicians = opdf.groupby('Patient_ID').apply(
    lambda group: pd.concat([
        group['Primary_Physician'],
        group['Operating_Physician'],
        group['Other_Physician']
    ]).nunique()
)

# Convert to dataframe
total_physicians_df = unique_physicians.reset_index()
total_physicians_df.columns = ['Patient_ID', 'Total_Physicians']

print(total_physicians_df)
```

# Total unique diagnoses

```python
# Combine all diagnosis columns to calculate total unique diagnoses per patient
total_diagnoses = opdf.groupby('Patient_ID').apply(
    lambda group: pd.concat([
        group['ICD9_DGNS_CD_1'], group['ICD9_DGNS_CD_2'], group['ICD9_DGNS_CD_3'], 
        group['ICD9_DGNS_CD_4'], group['ICD9_DGNS_CD_5'], group['ICD9_DGNS_CD_6'], 
        group['ICD9_DGNS_CD_7'], group['ICD9_DGNS_CD_8'], group['ICD9_DGNS_CD_9'], 
        group['ICD9_DGNS_CD_10']
    ]).nunique()
)

# Convert to dataframe
total_diagnoses_df = unique_diagnoses.reset_index()
total_diagnoses_df.columns = ['Patient_ID', 'Total_Diagnosis']

print(total_diagnoses_df)
```

# Most frequent diagnosis

```python
def most_freq_diagnosis(group):
    # Combine all diagnosis columns for a patient into a single list
    all_diagnoses = pd.concat([
        group['ICD9_DGNS_CD_1'], group['ICD9_DGNS_CD_2'], group['ICD9_DGNS_CD_3'], 
        group['ICD9_DGNS_CD_4'], group['ICD9_DGNS_CD_5'], group['ICD9_DGNS_CD_6'], 
        group['ICD9_DGNS_CD_7'], group['ICD9_DGNS_CD_8'], group['ICD9_DGNS_CD_9'], 
        group['ICD9_DGNS_CD_10']
    ])
    # Remove null values
    all_diagnoses = all_diagnoses.dropna()
    # Calculate the most frequent diagnosis
    if all_diagnoses.empty:
        return None
    return all_diagnoses.mode().iloc[0]  # Return the most frequent diagnosis

# Convert to dataframe
most_freq_diagnosis_df = opdf.groupby('Patient_ID').apply(most_freq_diagnosis).reset_index()
most_freq_diagnosis_df.columns = ['Patient_ID', 'Most_Freq_Diagnosis']

print(most_freq_diagnosis_df)
```

# Summarize Data

```python
# Merge all dataframes
summary_df = total_visits_df.merge(total_physicians_df, on='Patient_ID').merge(total_diagnoses_df, on='Patient_ID').merge(most_freq_diagnosis_df, on='Patient_ID')

print(summary_df)
```
